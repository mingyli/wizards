```python
import pycosat
import json

def linearizedIndex(numWizards, x, y):
    if (x<y):
        return -((numWizards**y)+x)
    return (numWizards**x)+y

def pprint(x):
    for c in x:
        print(c)

numWizards = 20

with open("failure-Andrew-20-0.txt") as data_file:
            data = json.load(data_file)
constraints=data[1]
cnf=[]

for c in constraints:
    a = linearizedIndex(numWizards, c[0], c[2])
    b = linearizedIndex(numWizards, c[1], c[2])
    cnf.append([a,-b])
    cnf.append([-a,b])
    
print(len(constraints))
pprint(cnf)
pprint(constraints)
#pycosat.solve(cnf)

#print(linearizedIndex(20,4,5))
#print(linearizedIndex(20,5,4))
```
